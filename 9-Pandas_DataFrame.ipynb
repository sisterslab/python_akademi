```python
df = pd.read_csv("movies.csv", usecols=['original_language', 'original_title', 'popularity',
       'release_date', 'vote_average', 'vote_count', 'genre', 'overview',
       'revenue', 'runtime', 'tagline'], engine="python")
```

```python
pd.isnull(df.runtime).sum()
```

```python
df.runtime.fillna(df.runtime.mean(), inplace=True)
```

```python
pd.isnull(df.runtime).sum()
```

```python
pd.isnull(df.popularity).sum()
```

```python
df.popularity.fillna(0, inplace=True)
```

```python
df.dropna(subset=["genre"], axis=0, inplace=True)
```

```python
df[df.genre.str.contains("Documentary")]
```

```python
df["upper_title"] = df.original_title.str.upper()

df[["original_title", "upper_title"]]
```

```python
import string

def remove_punc(text):
    title = text.translate(str.maketrans('', '',string.punctuation))
    return title

df["remove_punc"] = df.original_title.apply(remove_punc)

df[["original_title", "upper_title", "remove_punc"]]
```
