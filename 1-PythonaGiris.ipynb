**Yerel Değişken**

Tanımlandığı yerdeki kod bloğunda çalışan değişkenlerdir. Başka yerden erişilemezler.

```python
def getSalary1():
  salary1 = 10000
  print("Salary1 ", salary1)

def getSalary2():
  print("salary2 ", salary1)

getSalary1()
getSalary2()
```

**Global Değişkenler**

Kodun her yerinden erişilebilen değişkenlerdir.

```python
salary1 = 10000

def getSalary1():
  print("Salary1 ", salary1)

def getSalary2():
  print("salary2 ", salary1)

getSalary2()
getSalary1()

print("main ", salary1)
```

Global salary değişkeninin değerini değiştirelim

```python
salary1 = 10000

def getSalary1():
  salary1 = 15000
  print("Salary1 ", salary1)

def getSalary2():
  print("salary2 ", salary1)

getSalary1()
getSalary2()

print("main ", salary1)
```

```python
salary1 = 10000

def getSalary1():
  global salary1
  salary1 = 15000
  print("Salary1 ", salary1)

def getSalary2():
  print("salary2 ", salary1)

getSalary2()
getSalary1()

print("main ", salary1)
```

**Atama Operatörleri**

=, +=, -=, *=, /=, //=, %=

**=

```python
x = 1
print(x)

x += 2 # x = x + 2
print(x)

x -= 1 # x = x - 1
print(x)

x *= 9 # x = x * 9
print(x)

x /= 2 # x = x / 2
print(x)

x //= 4 # x = x // 4
print(x)

x **= 3 # x = x ** 3
print(x)

x %= 3 # x = x % 3
print(x)
```
